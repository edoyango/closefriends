```python
import seaborn as sb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Tests:
* data distributions: random square, ordered square, circular, 2-cluster, 4-cluster, jittered diagonal,
    * dependant: problem size + dims. Fixed average pairs per point
    * dependant: problem size + pairs per point. Fixed dims

## Random square, changing problem size and dims, fixed average pairs per point

```python
df_sp = pd.read_csv("times_sp_square.csv", index_col=0)
df_cf = pd.read_csv("times_cf_square.csv", index_col=0)
df_speed = df_sp/df_cf
fig = plt.figure(figsize=(7.5,3))

ax = sb.heatmap(df_speed, 
                annot=True, 
                mask=False,
                yticklabels=True,
                xticklabels=True,
                linewidths=0.5, 
                vmin=0,
                vmax=4,
                cbar_kws={"label": "t_scipy/t_closefriends"})
ax.xaxis.tick_top()
ax.set(xlabel="npoints", ylabel="dims")
lbls = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
```

## ndims hyperspheres, changing problem size and dims, fixed average pairs per point

```python
df_sp = pd.read_csv("times_sp_hyperspheres.csv", index_col=0)
df_cf = pd.read_csv("times_cf_hyperspheres.csv", index_col=0)
df_speed = df_sp/df_cf
fig = plt.figure(figsize=(7.5,3))

ax = sb.heatmap(df_speed, 
                annot=True, 
                mask=False,
                yticklabels=True,
                xticklabels=True,
                linewidths=0.5, 
                vmin=0,
                vmax=4,
                cbar_kws={"label": "t_scipy/t_closefriends"})
ax.xaxis.tick_top()
ax.set(xlabel="npoints", ylabel="dims")
lbls = ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```
